```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x, y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x2, y2, "b-.")

plt.plot(x, y)
plt.show()
```

```python
# color, marker, linewidth, legends, grid, x label, y label, title, text

# plot function x^3
import numpy as np
#x = np.array([1,2,3,4])
x = np.arange(0, 5, 0.1)
y  = x **3
y2 = x **2

plt.plot(x, y, "b", label='X^3')
plt.plot(x,y2,color="red", label = 'X^2')

# plt.axis([0, 10, 0, 200])

plt.grid()

plt.text(3-0.5, 3 **3, '3, 37', fontsize=10)

plt.ylabel('X^2 or  X^3')
plt.xlabel('X')

plt.legend()  # calling the legend function to plot legends
plt.title("Matplotlib Demo")

plt.show()
```

```python
a = [3,4,5]
b = [3,6,7]
plt.plot(a, color='blue', marker='o')
plt.plot(b, color='black',marker='*', linewidth=0.5)
plt.xlabel('X axis')
plt.ylabel("Y axis")
plt.show()
```

PI Graphs

```python
plt.title("Split Among Classes")

labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']

colors = ['red', 'green', 'blue', 'yellow']
sizes = [2,4,7,9]

explode = [0.1, 0, 0, 0.05]

plt.pie(sizes,colors=colors, labels=labels, autopct='%0.2f',explode=explode, shadow=True, counterclock=True, startangle=1000)  
# show percentage upto 2 decimals for each section use %0.2f%% to print % symbol
# counterclock parameter is for the explode property

# plt.axis('equal')  # to make graph circular from elliptical

plt.show()
```
